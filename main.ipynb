```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].progress_apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].progress_apply(convert)
```

```python
movies['keywords']
```

```python
movies.head()
```

```python
def find_name(text):
    L = []
    counter = 0
    
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].progress_apply(find_name)
```

```python
movies['cast']
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].progress_apply(fetch_director)
```

```python
movies['crew']
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(' ',''))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].progress_apply(lambda x: x.split())
```

```python
movies['overview']
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].progress_apply(lambda x: ' '.join(x))
```

```python
new['tags'] = new['tags'].progress_apply(lambda x: x.lower())
```

## Vectorization

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
        
```

```python
new['tags'] = new['tags'].progress_apply(stem)
```

```python
new['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=25000, stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```
